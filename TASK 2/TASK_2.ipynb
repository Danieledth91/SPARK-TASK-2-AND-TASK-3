**COUNT THE NUMBER OF ALL WORDS**

```python
from pyspark import SparkContext
from operator import add

if __name__ == "__main__":
    # Check if a SparkContext already exists
    try:
        sc = SparkContext.getOrCreate()
    except Exception as e:
        print("Error:", e)
        exit(1)

    print("SparkContext created successfully.")

    # Define the path to your text file
    data_path = "/content/drive/MyDrive/TASK 2/t8.shakespeare.txt"
    print("Data path:", data_path)

    # Read the text file
    lines = sc.textFile(data_path)
    print("Text file read successfully.")

    # Convert all letters to lowercase and perform word count
    counts = lines.flatMap(lambda x: x.lower().split(' ')).map(lambda x: (x, 1)).reduceByKey(add)
    print("Word count performed.")

    # Sort the word counts by word
    output = counts.sortBy(lambda x: x[0]).collect()
    print("Word counts sorted.")

    # Write the word counts to a file
    output_file_path = "Count_the_number_of_all_words.txt"
    with open(output_file_path, "w") as f:
        for (word, count) in output:
            f.write("%s: %i\n" % (word, count))
    print("Word counts written to file.")

    # Display the content of the written file at the console
    print("Content of", output_file_path, ":")
    with open(output_file_path, "r") as f:
        for line in f:
            print(line.strip())

    # Stop the SparkContext if it was created in this script
    if 'sc' in locals():
        sc.stop()
    print("SparkContext stopped.")
```

**REMOVE SPECIAL SYMBOLS**

```python
from pyspark import SparkContext
from operator import add
import re

if __name__ == "__main__":
    # Create or get SparkContext
    sc = SparkContext.getOrCreate()

    # Define the path to your text file
    data_path = "/content/drive/MyDrive/TASK 2/t8.shakespeare.txt"

    # Read the text file
    lines = sc.textFile(data_path, 1)

    # Modify the regular expression pattern to remove special symbols and count words
    counts = lines.flatMap(lambda x: re.findall(r'\b[a-zA-Z]+\b', x.lower())).map(lambda x: (x, 1)).reduceByKey(add)

    # Sort the word counts by word
    output = counts.sortBy(lambda x: x[0]).collect()

    # Write the word counts to a file
    with open("Remove_special_symbols.txt", "w") as f:
        for (word, count) in output:
            f.write("%s: %i\n" % (word, count))

    # Display the word counts at the console
    print("Word counts:")
    for (word, count) in output:
        print("%s: %i" % (word, count))

    # Stop the SparkContext
    sc.stop()
```

**OUTPUT 10 MOST POPULAR WORDS**

```python
from pyspark import SparkContext
from operator import add
import re

if __name__ == "__main__":
    # Define the path to your text file
    data_path = "/content/drive/MyDrive/TASK 2/t8.shakespeare.txt"

    # Initialize SparkContext
    sc = SparkContext.getOrCreate()

    # Read the text file
    lines = sc.textFile(data_path, 1)

    # Count words ignoring special symbols and converting to lowercase
    counts = lines.flatMap(lambda x: re.findall(r'\b[a-zA-Z]+\b', x.lower())).map(lambda x: (x, 1)).reduceByKey(add)

    # Sort word counts in descending order and take top 10
    sorted_counts = counts.sortBy(lambda x: x[1], ascending=False)
    top_10 = sorted_counts.take(10)

    # Write top 10 words to a file
    with open("top_10_words.txt", "w") as f:
        for (word, count) in top_10:
            f.write("%s: %i\n" % (word, count))

    # Display the top 10 words at the console
    print("Top 10 words:")
    for (word, count) in top_10:
        print("%s: %i" % (word, count))

    # Stop SparkContext
    sc.stop()
```
